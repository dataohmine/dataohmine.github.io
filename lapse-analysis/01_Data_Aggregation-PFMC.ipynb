```python
filenames = glob.glob(path + "\*events_monthly*.csv")
print(filenames)
```

### Join PFMC Inf and Events to get Lapse policiies  (3month ahead)

```python
# Merge pfmv and events
df = pfmc.merge(events[['agmt_pkge_id','evnt_desc','lapse_ind', 'surrender_ind', 'reinstate_ind',
                        'evnt_1_month_ago', 'evnt_2_month_ago', 'evnt_3_month_ago', 
                        'evnt_cur_month']], 
              how='left', 
              left_on=['agmt_pkge_id', 'pfmc_cur_month'], 
              right_on=['agmt_pkge_id', 'evnt_3_month_ago'], 
              suffixes=('', '_remove'))

df.drop([i for i in df.columns if 'remove' in i],axis=1, inplace=True)
```

```python
## Lagged lapse indicators
df['lapse_ind']= df['lapse_ind'].fillna(0)
df['lapse_ind_lag1'] = df.groupby('agmt_pkge_id')['lapse_ind'].shift(1)
df['lapse_ind_lag1']=df['lapse_ind_lag1'].fillna(0)

df['lapse_ind_lag2'] = df.groupby('agmt_pkge_id')['lapse_ind'].shift(2)
df['lapse_ind_lag2']=df['lapse_ind_lag2'].fillna(0)

df['lapse_ind_lag3'] = df.groupby('agmt_pkge_id')['lapse_ind'].shift(3)
df['lapse_ind_lag3']=df['lapse_ind_lag3'].fillna(0)
```

```python
df['reinstate_ind']= df['reinstate_ind'].fillna(0)
df['reinstate_ind_lag1'] = df.groupby('agmt_pkge_id')['reinstate_ind'].shift(1)
df['reinstate_ind_lag1']=df['reinstate_ind_lag1'].fillna(0)

df['reinstate_ind_lag2'] = df.groupby('agmt_pkge_id')['reinstate_ind'].shift(2)
df['reinstate_ind_lag2']=df['reinstate_ind_lag2'].fillna(0)

df['reinstate_ind_lag3'] = df.groupby('agmt_pkge_id')['reinstate_ind'].shift(3)
df['reinstate_ind_lag3']=df['reinstate_ind_lag3'].fillna(0)
```

```python
df['surrender_ind']= df['surrender_ind'].fillna(0)
df['surrender_ind_lag1'] = df.groupby('agmt_pkge_id')['surrender_ind'].shift(1)
df['surrender_ind_lag1']=df['surrender_ind_lag1'].fillna(0)

df['surrender_ind_lag2'] = df.groupby('agmt_pkge_id')['surrender_ind'].shift(2)
df['surrender_ind_lag2']=df['surrender_ind_lag2'].fillna(0)

df['surrender_ind_lag3'] = df.groupby('agmt_pkge_id')['surrender_ind'].shift(3)
df['surrender_ind_lag3']=df['surrender_ind_lag3'].fillna(0)
```

```python
df['cur_status']= np.where(df['lapse_ind']==1, 'lapse', 
                           (np.where(df['surrender_ind']==1, 'surrender', 
                             (np.where(df['reinstate_ind']==1, 'reinstated',  "current")))))
```

```python
df['prior_1mo_status']= np.where(df['lapse_ind_lag1']==1, 'lapse', 
                           (np.where(df['surrender_ind_lag1']==1, 'surrender', 
                           (np.where(df['reinstate_ind_lag1']==1, 'reinstated',  "current")))))
```

```python
df['prior_2mo_status']= np.where(df['lapse_ind_lag2']==1, 'lapse', 
                           (np.where(df['surrender_ind_lag2']==1, 'surrender', 
                           (np.where(df['reinstate_ind_lag2']==1, 'reinstated',  "current")))))
```

```python
df['prior_3mo_status']= np.where(df['lapse_ind_lag3']==1, 'lapse', 
                           (np.where(df['surrender_ind_lag3']==1, 'surrender', 
                           (np.where(df['reinstate_ind_lag3']==1, 'reinstated',  "current")))))
```
